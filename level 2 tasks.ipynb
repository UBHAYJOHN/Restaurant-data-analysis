```python
df=pd.read_csv('cleaned_task1.csv')
df.head(5)
```

```python
from itertools import combinations
```

```python
from collections import Counter
```

```python
df=df.dropna(subset=['Cuisines'])
```

```python
df['Cuisines']=df['Cuisines'].apply(lambda x: [i.strip() for i in x.split(',')])
```

```python
pair_counter=Counter()
```

```python
for Cuisines in df['Cuisines']:
    if len(Cuisines)>=2:
        pair=combinations(sorted(set(Cuisines)),2)
        pair_counter.update(pair)
```

```python
combo_df=pd.DataFrame(pair_counter.items(),columns=['cuisines_pairs','pair'])
```

```python
combo_df = combo_df.sort_values(by='pair', ascending=False)
```

```python
print(combo_df.columns)
```

#task 3 

Determine if certain cuisine combinations tend to have higher ratings

```python
from collections import defaultdict
```

```python
df.head(10)
```

```python
df=df.dropna(subset=['Aggregate rating'])
```

```python
df.tail(10)
```

```python
pair_ratings=defaultdict(list)
```

```python
for _,row in df.iterrows():
    cuisines=sorted(set(rows['Cuisines']))
    rating=row['Aggregate rating']
    if len(cuisines)>=2:
        pairs=combinations(cuisines,2)
        for pair in pairs:
            pair_ratings[pair].append(rating)
```

```python
# now we need to use the avg of rating ,the reason is  it normalizes multiple entries per pair
```

```python
avg_rating_data = [(pair, sum(ratings) / len(ratings)) for pair, ratings in pair_ratings.items()]
rating_df = pd.DataFrame(avg_rating_data, columns=['Cuisine_Pair', 'Average_Rating'])
```

```python
print(rating_df)
```

```python
rating_df=rating_df.sort_values(by='Average_Rating',ascending=False)
```

```python
print(rating_df)
```

```python
high_rating_combos = rating_df[rating_df['Average_Rating'] > 4.0]
print(high_rating_combos)
```
